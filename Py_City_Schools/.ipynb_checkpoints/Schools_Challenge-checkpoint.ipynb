```python
# Merge Tables 2 DF's together based on school they share

merge_df = pd.merge(schools_df, students_df, how="left", on=["school_name", "school_name"])
merge_df.head(),
```

```python
#District Summary 1
#Create a high level snapshot (in table form) of the district's key metrics, including:

# Create List of Schools 
Total_Schools_df = merge_df["school_name"].value_counts()
print (Total_Schools_df)

# Collecting and Printing total School Count
School_Count = len(schools_df["school_name"].unique())
School_Count
print ("School Count", School_Count)

# Collecting and Printing total students Count
Student_Count = len(students_df["student_name"])
print ("Student Count", Student_Count)

# Calculating and Printing total Budget
Budget_Sum = sum(schools_df["budget"])
print ("Budget Sum", Budget_Sum)

# Calculating and Printing Avg Reading Score
Avg_Read = students_df["reading_score"].mean()
print ("Avg Reading Score", Avg_Read)
    
# Calculating and Printing Avg Math Score
Avg_Math = students_df["math_score"].mean()
print ("Avg Math Score", Avg_Math)    

# Calculating and Printing Avg Math Pass Percentage
Math_Pass = students_df.loc[students_df["math_score"]>= 70, ["math_score"]]
Math_Pct = Math_Pass["math_score"].count()/Student_Count *100
print ("Pct Passing Math", Math_Pct)   

# Calculating and Printing Avg Reading Pass Percentage
Read_Pass = students_df.loc[students_df["reading_score"]>= 70, ["reading_score"]]
Read_Pct = Read_Pass["reading_score"].count()/Student_Count *100
print ("Pct Passing Reading", Read_Pct)   

# Calculating and Printing Avg Pass Percentage
# composite score
comp_scr = ((students_df["reading_score"].count() + students_df["math_score"].count())/2)/Student_Count *100
print ("Comp_Score", comp_scr)

Over_Pass = (Read_Pct + Math_Pct) /2
print ("Overall Pct Passing", Over_Pass)   
```
